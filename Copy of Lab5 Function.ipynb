1 เขียนฟังก์ชัน greet_user(name) ที่รับพารามิเตอร์ชื่อ name และพิมพ์ข้อความทักทายผู้ใช้ตามที่นักศึกษาต้องการ พร้อมทั้งเรียกใช้ฟังก์ชัน
```
# ตัวอย่างการเรียกใช้ฟังก์ชัน
greet_user("Alice")  
# ผลลัพธ์:
Sawasdee ka, Alice!
```


````python
#ประกาศฟังก์ชัน
def  greet_user(name):
  print(f"Sawasdee ka, {name}")
````

````python
#เรียกใช้ฟังก์ชัน
greet_user("Alice!")
````

3 สร้างฟังก์ชัน isGreaterThan(a, b) แบบไม่ return ค่า ที่รับเลขจำนวนเต็มสองตัวและแสดงผลตัวเลขที่มีค่ามากกว่า

5 เขียนฟังก์ชันแบบ return ค่ากลับ ชื่อ calculate_interest(principal, rate, time) ที่คำนวณดอกเบี้ยแบบง่ายตามสูตร:



Screenshot 2024-09-17 170214.png


```
#ตัวอย่างการเรียกใช้
interest = calculate_interest(1000, 5, 2)
print(interest)  
# ผลลัพธ์:
100.0
```


````python
#ประกาศฟังก์ชัน
def calculate_interest(principal, rate, time):
  interest = principal * rate * time / 100
  return interest
````

````python
#เรียกใช้ฟังก์ชัน
interest = calculate_interest(1000, 5, 2)
print(interest)
````

6 เขียนฟังก์ชัน is_even(number) ที่รับตัวเลขและคืนค่า True ถ้าตัวเลขเป็นเลขคู่ และคืนค่า False ถ้าตัวเลขเป็นเลขคี่



```
# ตัวอย่าง
print(is_even(4))
print(is_even(7))  
# ผลลัพธ์:
True
False
```



````python
#ประกาศฟังก์ชัน
def is_even(a):
  if a % 2 == 0:
    return True
  else:
    return False
````

````python
#เรียกใช้ฟังก์ชัน
print(is_even(4))
print(is_even(7))
````

7 เขียนโปรแกรมแปลงค่าอุณหภูมิจาก Farenheit เป็น Celcius (ศึกษาสูตรจากเนตนะจ้ะ)


```
#ตัวอย่างการเรียกใช้
f = 98.6
c = fahrenheit_to_celsius(f)
print(f"{f} Fahrenheit = {c:.2f} Celsius")
#ผลลัพธ์
98.6 Fahrenheit = 37.00 Celsius
```



````python
#ประกาศฟังก์ชัน
def fahrenheit_to_celsius(f):
  c = ((f - 32) * 5) / 9
  return c
````

````python
#เรียกใช้ฟังก์ชัน
f = 98.6
c = fahrenheit_to_celsius(f)
print(f"{f} Fahrenheit = {c:.2f} Celsius")
````

8 เขียนฟังก์ชันแสดงแม่สูตรคูณจากตัวเลขที่กำหนด

```
#ตัวอย่างการเรียกใช้
multiplication_table(3)
#ผลลัพธ์
3 x 1 = 3
3 x 2 = 6
3 x 3 = 9
3 x 4 = 12
.
.
.
3 x 12 = 36
```



````python
#ประกาศฟังก์ชัน
def multiplication_table(a):
  for i in range(1,13):
    print(f"{a} x {i} = {a * i}")
````

````python
#เรียกใช้ฟังก์ชัน
multiplication_table(3)
````
