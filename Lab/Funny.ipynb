```python
Account = {}

def Create_Account(name, balance, password):
  global Account
  Account.update({
    name : {
        "balance" : balance,
        "password" : password
    }
  })
  print(f'\n========== สร้างสำเร็จ ==========\nName : {name}\nBalance {Account[name]["balance"]}\nPassword : {Account[name]["password"]}\n=============================\n')

def Check(check, username, password):
  global Account
  if check == "Password":
    if Account[username]["password"] == password:
      print(f"\nรหัสถูกต้อง ยินดีต้อนรับคุณ {username}")
      return True
    else:
      print("\nXXXXXX รหัสผ่านไม่ถูกต้อง XXXXXX\n")
      return False
  elif check == "Account":
    if Account == {}:
      return False
    else:
      return True
  elif check == "NameAccount":
    if username not in Account:
      print("\nXXXXXX ไม่มีชื่อบัญชีนี้ในระบบ XXXXXX\n")
      return False
    else:
      return True


def Show():
  global Account
  for user in Account:
    print(f'\n================================\nName : {user}\nBalance {Account[user]["balance"]}\nPassword : {Account[user]["password"]}\n================================\n')

def getBalance(username, password):
  global Account
  if Check("Password", username, password) != True:
    return None
  else:
    print(f'\n-_-_- คุณมีเงินในบัญชี : {Account[username]["balance"]} -_-_-\n')
    return Account[username]["balance"]

def Deposits(username, password, amount):
  global Account
  if Check("Password", username, password) != True:
    return None
  if amount <= 0:
    print("XXX กรุณาระบุจำนวนเงินที่ถูกต้อง XXX")
  else:
    Account[username]["balance"] += amount
    print(f'>>>> ฝากเงินจำนวน : {amount} ยอดคงเหลือในบัญชี : {Account[username]["balance"]} <<<<')
    return None

def Withdraw(username, password, amount):
  global Account
  if Check("Password", username, password) != True:
    return None
  if amount <= Account[username]["balance"]:
    Account[username]["balance"] -= amount
    print(f'>>>> ถอนเงินจำนวน : {amount} ยอดคงเหลือในบัญชี : {Account[username]["balance"]} <<<<')
    return None
  else:
    print("XXX กรุณาระบุจำนวนเงินที่ถูกต้อง ยอดเงินในบัญชีคุณไม่เพียงพอ XXX")
    return None

def ActionMain(Key_Input):
  if Key_Input == "a":
    print("\n<<<< สร้างบัญชีธนาคาร >>>>")
    username = input("\rกรุณาใส่ชื่อบัญชี : ")
    if username.strip() == "": return print("\nXXXXX กรุณาใส่ชื่อบัญชี XXXXX\n")
    password = input("\rกรุณาใส่รหัสผ่านบัญชี : ")
    if password.strip() == "": return print("\nXXXXX กรุณาใส่รหัสผ่าน XXXXX\n")
    Create_Account(username, 0, password)
  elif Key_Input == "b":
    username = input("\rกรุณาใส่ชื่อบัญชี : ")
    if Check("NameAccount", username, "") is False: return None
    userPassword = input("\rกรุณากรอกรหัสผ่าน : ")
    getBalance(username, userPassword)
  elif Key_Input == "d":
    username = input("\rกรุณาใส่ชื่อบัญชี : ")
    if Check("NameAccount", username, "") is False: return None
    userPassword = input("กรุณากรอกรหัสผ่าน : ")
    userDepositAmount = int(input("\nใส่จำนวนเงินที่จะฝาก "))
    Deposits(username, userPassword, userDepositAmount)
  elif Key_Input == "w":
    username = input("\rกรุณาใส่ชื่อบัญชี : ")
    if Check("NameAccount", username, "") is False: return None
    userPassword = input("กรุณากรอกรหัสผ่าน : ")
    userWithdrawAmount = int(input("\rใส่จำนวนเงินที่จะถอน "))
    Withdraw(username, userPassword, userWithdrawAmount)
  elif Key_Input == "s":
    Show()
  else :
    print("\nXXXXX กรุณาพิมพ์ให้ถูกต้อง XXXXX\n")
```

```python
while True:
  while Check("Account","","") is False:
    try:
      print("\nXXXXX ยังไม่มีบัญชีในระบบกรุณากด A เพื่อสร้างบัญชี XXXXX")
      action = input(">> ").lower()[0]
    except IndexError:
      pass
    if action == "a":
      ActionMain(action)
  else:
    print("กด A สร้างบัญชี\nกด B ดูเงิน\nกด D ฝากเงิน\nกด W ถอนเงิน\nกด S โปรไฟล์ทั้งหมด\nกด Q ออกจากระบบ")
    try:
      action = input(">> ").lower()[0]
    except IndexError:
      pass
    if action == "q":
      print("\nXXXXX ปิดโปรแกรม XXXXX\n")
      break
    ActionMain(action)
```
