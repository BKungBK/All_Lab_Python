1. เขียนโปรแกรมเพื่อรับค่าตัวแปร a และ b ชนิดจำนวนเต็มจากผู้ใช้ จากนั้นแสดงผลลัพธ์ของการเปรียบเทียบค่าด้วยเครื่องหมาย >, <, >=, <=, ==, และ !=

  ตัวอย่างการรันโปรแกรม<br>
  ```
    Enter a: 10
    Enter b: 20
    a<b  is True
    a>b  is False
    a<=b is True
    a>=b is False
    a==b is False
    a!=b is True

  ```




````python
a = int(input("Enter a: "))
b = int(input("Enter b: "))

print(f"a < b is {a < b}\na > b is {a > b}\na <= b is {a <= b}\na >= b is {a >= b}\na == b is {a == b}\na != b is {a != b}")
````

7. เขียนโปรแกรมรับอินพุตที่เป็นจำนวนเต็ม 1 จำนวนแล้วตรวจสอบข้อมูลนำเข้าว่าเป็นเลขคู่หรือเลขคี่<br>
หากเป็นเลขคู่แสดงผล This is an even number.<br>
หากเป็นเลขคี่แสดงผล This is an odd number.

  ตัวอย่าง <br>

  ```
      Enter any number: 2
      This is an even number.
  ```
    ```
      Enter any number: 1
      This is an odd number.
  ```
  คำใบ้: ใช้เครื่องหมาย % ในการเขียนเงื่อนไข

````python
num = int(input("Enter any number: "))

if num % 2 == 0:
  print("This is an even number.")
else:
  print("This is an odd number.")
````

8. เขียนโปรแกรมแสดงการคำนวณการเสียภาษี ตามเงื่อนไขนี้
ป้อนข้อมูลรายได้

  คำนวณภาษี ดังนี้<br>
  รายได้น้อยกว่าเท่ากับ 15000 บาท ไม่เสียภาษี (No need to pay tax)<br>
  รายได้15001 – 25000 บาท เสียภาษี 10 % (Your tax amount is …)<br>
  รายได้มากกว่า 25000 ขึ้นไป เสีย 15 % (Your tax amount is …)<br>
  แสดงผลลัพธ์จำนวนภาษีที่ต้องจ่าย ตามวงเล็บที่ระบุข้างต้น<br>

  ตัวอย่าง<br>

  ```
    Enter salary: 30000
    Your tax amount is 4500 B
  ```
  ```
    Enter salary: 10000
    No need to pay tax
  ```



````python
salary = int(input("Enter salary: "))

if salary <= 15000:
  print("No need to pay tax")
elif salary > 15000 and salary < 25000:
  salary = (salary // 100) * 10
  print(f"Your tax amount is {salary} B.")
else:
  salary = (salary // 100) * 15
  print(f"Your tax amount is {salary} B.")
````

9. เขียนโปรแกรมเพื่อรับค่าคะแนนรวมวิชาการโปรแกรมคอมพิวเตอร์ ของนักศึกษา(เต็ม100) จากนั้นแสดงผลเกรดที่ได้<br>
หากคะแนนอยู่ในช่วง 80 - 100 แสดงผล Grade A<br>
หากคะแนนอยู่ในช่วง 70 - 79  แสดงผล Grade B<br>
หากคะแนนอยู่ในช่วง 60 - 69  แสดงผล Grade C<br>
หากคะแนนอยู่ในช่วง 50 - 59  แสดงผล Grade D<br>
หากคะแนนอยู่ในช่วง 0  - 50  แสดงผล Grade F<br>
หากคะแนนอยู่นอกเหนือจากช่วงข้างต้น แสดงข้อความ Invalid score<br>

  ตัวอย่าง
  ```
  Enter Computer Programming score: 90
  Grade A
  ```
  ```
  Enter Computer Programming score: 65
  Grade C
  ```
  ```
  Enter Computer Programming score: -5
  Invalid score
  ```

````python
sc = int(input("Enter Computer Programming score: "))

if sc >= 80 and sc <= 100:
  print("Grade A")
elif sc >= 70 and sc <= 79:
  print("Grade B")
if sc >= 60 and sc <= 69:
  print("Grade C")
elif sc >= 50 and sc <= 59:
  print("Grade D")
elif sc >= 0 and sc <= 49:
  print("Grade F")
else:
  print("Invalid score")
````
