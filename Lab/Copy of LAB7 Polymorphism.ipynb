#Part 1 : Polymorphism
1. สร้างคลาส Circle, Rectangle, และ Triangle โดยแต่ละคลาสจะต้องเก็บตัวแปรที่ใช้ในการคำนวณพื้นที่ของแต่ละรูปทรง และมีเมธอด calculate_area() ซึ่งคำนวณพื้นที่ของรูปทรงนั้น ๆ
2. สร้างคลาส Shapes โดยในคลาสจะประกอบไปด้วย

  *   List สำหรับเก็บรูปทรงต่างๆ
  *   เมธอด add_shape() เพื่อเพิ่มรูปทรงที่ป้อน ไปยัง List
  *   เมธอด print_area()เพื่อแสดงพื้นที่ของรูปทรงทั้งหมดใน List
  *   เมธอด total_area() เพื่อคำนวณพื้นที่รวมของรูปทรงทั้งหมดใน List

3. สร้างออปรูปทรงต่างๆ (Circle Rectangle และ Triangle) และเพิ่มไปยังออปเจคคลาส Shapes
4. คำนวณและแสดงผลพื้นที่ของแต่ละรูปทรงในคลาส Shapes
5. แสดงผลรวมของพื้นที่ทั้งหมดของรูปทรงในคลาส Shapes



````python
import math as m

class Circle:
  def __init__(self, r):
      self.__r = r

  def get_r(self):
      return self.__r

  def set_r(self, value):
      if value > 0:
          self.__r = value
      else:
          print("Radius must be positive!")

  def calculate_area(self):
    a = m.pi * self.__r**2
    return a

  def __str__(self):
    return "รูปทรง Circle"

class Rectangle:
  def __init__(self, w, h):
      self.__w = w
      self.__h = h

  def get_w(self):
      return self.__w

  def set_w(self, value):
      if value > 0:
          self.__w = value
      else:
          print("Width must be positive!")

  def get_h(self):
      return self.__h

  def set_h(self, value):
      if value > 0:
          self.__h = value
      else:
          print("Height must be positive!")

  def calculate_area(self):
    a = self.__w * self.__h
    return a

  def __str__(self):
    return "รูปทรง Rectangle"

class Triangle:
  def __init__(self, b, h):
      self.__b = b
      self.__h = h

  def get_b(self):
      return self.__b

  def set_b(self, value):
      if value > 0:
          self.__b = value
      else:
          print("Base must be positive!")

  def get_h(self):
      return self.__h

  def set_h(self, value):
      if value > 0:
          self.__h = value
      else:
          print("Height must be positive!")

  def calculate_area(self):
    a = (1/2) * self.__b * self.__h
    return a

  def __str__(self):
    return "รูปทรง Triangle"

class Shapes:
  def __init__(self):
      self.__shapes_list = []

  def get_shapes_list(self):
      return self.__shapes_list

  def set_shapes_list(self, value):
      if isinstance(value, list):
          self.__shapes_list = value
      else:
          print("Shapes list must be a list!")

  def add_shape(self, shape):
    self.__shapes_list.append(shape)

  def print_area(self):
    print("| รูปทรงทั้งหมด |")
    for i in self.__shapes_list:
      print(i, end=" | ")

  def total_area(self):
    for i in self.__shapes_list:
      print(f"{i} = {i.calculate_area()}")

  def All(self):
    a = 0
    for i in self.__shapes_list:
      a += i.calculate_area()
    print(f"\n| ผลรวมทั้งหมด : {a} |")

s = Shapes()
s.add_shape(Circle(20))
s.add_shape(Circle(10))
s.add_shape(Rectangle(20, 20))
s.add_shape(Rectangle(10, 20))
s.add_shape(Triangle(15, 20))
s.add_shape(Triangle(10, 15))
s.total_area()
s.print_area()
s.All()
````

# Part 2: Operator Overloading
2.1 ให้นักศึกษาจะสร้างคลาส Money เพื่อจัดการกับจำนวนเงินในสกุลเงินบาท โดยเงินจะถูกแบ่งออกเป็น "บาท" และ "สตางค์" (1 บาท = 100 สตางค์) โดยโปรแกรมต้องสามารถบวกจำนวนเงินสองจำนวนได้และแสดงผลลัพธ์ออกมาในรูปแบบที่ถูกต้องโดยใช้เครื่องหมาย +


ตัวอย่าง

```
m1 = Money(100, 75)
m2 = Money(50, 50)
print(m1 + m2)  # ฿151.25
```

1. สร้างคลาส Money โดยคลาสนี้ประกอบด้วย:

  * เมธอด `__init__` เพื่อกำหนดค่าเริ่มต้นของ "บาท" และ "สตางค์"
  * เมธอด `__add__` เพื่อบวกจำนวนเงินสองจำนวนและคำนวณการเปลี่ยนแปลงที่เกิดจากการรวมสตางค์เกิน 100
  * เมธอด `__str__` เพื่อแสดงผลลัพธ์ของจำนวนเงินในรูปแบบที่มีสัญลักษณ์ "฿" เช่น "฿<บาท>.<สตางค์สองหลัก>" (ให้ศึกษาการทำงานของเมธอด `__str__` เพิ่มเติมจากอินเตอร์เนต)
  
2. ทดสอบการทำงานโดยสร้างอ็อบเจ็กต์ Money ที่มีจำนวนเงินในรูปแบบของบาทและสตางค์ แล้วทำการบวกจำนวนเงินโดยใช้เครื่องหมาย + และแสดงผล



````python
class Money:
  def __init__(self, b, s):
      self.__b = b
      self.__s = s
  def get_b(self):
      return self.__b

  def set_b(self, value):
      if value >= 0:
          self.__b = value
      else:
          print("Balance must be non-negative!")

  def get_s(self):
      return self.__s

  def set_s(self, value):
      if value >= 0:
          self.__s = value
      else:
          print("Savings must be non-negative!")

  def __add__(self, other):
    b = self.__b + other.get_b()
    s = self.__s + other.get_s()
    while s >= 100:
      s -= 100
      b += 1
    return f"฿{b}.{s}"

  def __str__(self):
    return f"฿{self.__b}.{self.__s}"

m1 = Money(100, 75)
m2 = Money(50, 50)
print(m1 + m2)  # ฿151.25
print(m1)
````

2.2 สร้างคลาส Time เพื่อจัดการกับเวลา โดยเวลาในที่นี้จะถูกเก็บเป็นชั่วโมง นาที และวินาทีโดยต้องสามารถบวกและแสดงส่วนต่างของเวลาได้ และแสดงผลลัพธ์ในรูปแบบที่เหมาะสม ดังตัวอย่าง

  ```
  t1 = Time(5, 45, 30)
  t2 = Time(2, 30, 45)
  print(t1 + t2)  #8 hours 16 minutes 15 seconds
  print(t1 - t2)  #3 hours 14 minutes 45 seconds

  ```



1. สร้างคลาส Time โดยมีเมธอดต่างๆ ดังนี้:

    *  `__init__`: เมธอดเริ่มต้นที่รับค่าชั่วโมง นาที และวินาที
    *  `__add__`: เมธอดสำหรับการบวกเวลาสองชุด
    *  `__sub__`: เมธอดสำหรับการหาความแตกต่างระหว่างเวลาสองชุด
    *  `__str__`: เมธอดสำหรับการแสดงผลเวลาในรูปแบบ "x hours and y minutes z minutes"

    
2. สร้างอ็อบเจ็กต์ Time สองตัวที่มีเวลาต่างกัน เช่น:

  ```
      t1 = Time(5, 45, 30)
      t2 = Time(2, 30, 45)
  ```
3. ทำการบวกเวลา t1 และ t2 และแสดงผลลัพธ์

4. ทำการหาความแตกต่างระหว่างเวลา t1 และ t2 และแสดงผลลัพธ์  




````python
class Time:
  def __init__(self, h, m, s):
      self.__h = h
      self.__m = m
      self.__s = s

  def get_h(self):
      return self.__h

  def set_h(self, value):
      if 0 <= value < 24:
          self.__h = value
      else:
          print("Hour must be between 0 and 23!")

  def get_m(self):
      return self.__m

  def set_m(self, value):
      if 0 <= value < 60:
          self.__m = value
      else:
          print("Minute must be between 0 and 59!")

  def get_s(self):
      return self.__s

  def set_s(self, value):
      if 0 <= value < 60:
          self.__s = value
      else:
          print("Second must be between 0 and 59!")

  def __add__(self, other):
    h = self.__h + other.get_h()
    m = self.__m + other.get_m()
    s = self.__s + other.get_s()
    while m >= 60 or s >= 60:
      if s >= 60:
        m += 1
        s -= 60
      elif m >= 60:
        h += 1
        m -= 60
    return f"{h} hours {m} minutes {s} seconds"

  def __sub__(self, other):
    h = self.__h - other.get_h()
    m = self.__m - other.get_m()
    s = self.__s - other.get_s()
    while m < 0 or s < 0:
      if s < 0:
        m -= 1
        s = 60 - abs(s)
      elif m < 0:
        h -= 1
        m = 60 - abs(m)
    return f"{h} hours {m} minutes {s} seconds"

t1 = Time(5, 45, 30)
t2 = Time(2, 30, 45)
print(t1 + t2)  #8 hours 16 minutes 15 seconds
print(t1 - t2)  #3 hours 14 minutes 45 seconds
````
