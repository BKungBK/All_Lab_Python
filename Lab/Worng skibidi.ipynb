```python
Name_Account0 = ""
Balance_Account0 = 0
Password_Account0 = ""
Name_Account1 = ""
Balance_Account1 = 0
Password_Account1 = ""

def Account_Create(accnum, name, balance, password):
  global Name_Account0, Balance_Account0, Password_Account0
  global Name_Account1, Balance_Account1, Password_Account1
  if accnum == 0:
    Name_Account0 = name
    Balance_Account0 = balance
    Password_Account0 = password
  elif accnum == 1:
    Name_Account1 = name
    Balance_Account1 = balance
    Password_Account1 = password

def CheckPassword(accnum, password):
  global Password_Account0, Password_Account1
  if accnum == 0:
    if password != Password_Account0:
      print("รหัสไม่ถูกต้อง")
      return False
    else:
      return True
  elif accnum == 1:
    if password != Password_Account1:
      print("รหัสไม่ถูกต้อง")
      return False
    else:
      return True

def Show():
  global Name_Account0, Balance_Account0, Password_Account0
  global Name_Account1, Balance_Account1, Password_Account1
  print(f" Name : {Name_Account0}\n Balance : {Balance_Account0}\n Password : {Password_Account0}\n")
  print(f" Name : {Name_Account1}\n Balance : {Balance_Account1}\n Password : {Password_Account1}\n")

def getBalance(accnum, password):
  global Name_Account0, Balance_Account0, Password_Account0
  global Name_Account1, Balance_Account1, Password_Account1
  if CheckPassword(accnum, password) != True:
    return None
  if accnum == 0:
    return Balance_Account0
  elif accnum == 1:
    return Balance_Account1

def Deposits(accnum, amountDepoits, password):
  global Name_Account0, Balance_Account0, Password_Account0
  global Name_Account1, Balance_Account1, Password_Account1
  if amountDepoits <= 0:
    print("ใส่ค่าดีๆ")
    return None
  if CheckPassword(accnum, password) != True:
    return None
  if accnum == 0:
    Balance_Account0 = Balance_Account0 + amountDepoits
    return Balance_Account0
  elif accnum == 1:
    Balance_Account1 = Balance_Account1 + amountDepoits
    return Balance_Account1

def Withdraw(accnum, amountWithdraw, password):
  global Name_Account0, Balance_Account0, Password_Account0
  global Name_Account1, Balance_Account1, Password_Account1
  if amountWithdraw <= 0:
    print("ใส่ค่าดีๆ")
    return None
  if CheckPassword(accnum, password) != True:
    return None
  if accnum == 0:
    if Balance_Account0 == amountWithdraw:
      print("ถอนเงินทั้งหมด")
      return 0
    elif Balance_Account0 < amountWithdraw:
      print("เงินไม่พอให้ถอน")
      return None
  elif accnum == 1:
    if Balance_Account1 == amountWithdraw:
      print("ถอนเงินทั้งหมด")
      return 0
    elif Balance_Account1 < amountWithdraw:
      print("เงินไม่พอให้ถอน")
    return None
  if accnum == 0:
    Balance_Account0 = Balance_Account0 + amountWithdraw
    return Balance_Account0
  elif accnum == 1:
    Balance_Account1 = Balance_Account1 + amountWithdraw
    return Balance_Account1

Account_Create(0, "Skibidi", 500, "1234")
Account_Create(1, "NU", 5, "1234")
```

```python
while True:
  print("กด B ดูเงิน\nกด D ฝากเงิน\nกด W ถอนเงิน\nกด S โปรไฟล์\nกด Q ออกจากระบบ")
  action = input("ต้องการจะทำอะไร... \n> ").lower()[0]
  if action == "b":
    accnum = int(input("ใส่เลข account : "))
    userPassword = input("\nกรุณากรอกรหัสผ่าน : ")
    if getBalance(accnum, userPassword) is not None:
      print(f"คุณมีเงินอยู่จำนวน : {getBalance(accnum, userPassword)}\n")
  elif action == "d":
    userDepositAmount = int(input("\nใส่จำนวนเงินที่จะฝาก "))
    accnum = int(input("ใส่เลข account : "))
    userPassword = input("กรุณากรอกรหัสผ่าน : ")
    if Deposits(accnum, userDepositAmount, userPassword) is not None:
      print(f"เงินของคุณคงเหลือ : {Deposits(accnum, userDepositAmount, userPassword)}\n")
  elif action == "w":
    userWithdrawAmount = int(input("\nใส่จำนวนเงินที่จะถอน "))
    accnum = int(input("ใส่เลข account : "))
    userPassword = input("กรุณากรอกรหัสผ่าน : ")
    if Withdraw(accnum, userWithdrawAmount, userPassword) is not None:
      print(f"เงินของคุณคงเหลือ : {Withdraw(accnum, userWithdrawAmount, userPassword)}\n")
  elif action == "s":
    Show()
  elif action == "q":
    print("\nXXXXX ปิดโปรแกรม XXXXX\n")
    break
  else :
    print("\nXXXXX กรุณาพิมพ์ให้ถูกต้อง XXXXX\n")
```
