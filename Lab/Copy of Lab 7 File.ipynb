ใช้คำสั่ง open เพื่อสร้างไฟล์ dummy_file.txt โดยมีข้อความในไฟล์ดังต่อไปนี้


```
    The sun rises in the east every morning.
    Python is a versatile programming language.
    Artificial intelligence is shaping the future of technology.
    Reading books can expand your knowledge and creativity.
    Healthy eating and regular exercise improve well-being.

```



````python
#ไม่ใช้คำสั่ง with
dummy = open('dummy_file.txt', 'w')
dummy.write('The sun rises in the east every morning.\nPython is a versatile programming language.\nArtificial intelligence is shaping the future of technology.\nReading books can expand your knowledge and creativity.\nHealthy eating and regular exercise improve well-being.')
dummy.close()
````

````python
#ใช้คำสั่ง with
with open('dummy_file.txt', 'w') as dummy:
  dummy.write('The sun rises in the east every morning.\nPython is a versatile programming language.\nArtificial intelligence is shaping the future of technology.\nReading books can expand your knowledge and creativity.\nHealthy eating and regular exercise improve well-being.')
````

เขียนโค้ดเพื่อเปิดไฟล์และแสดงชื่อของไฟล์ dummy_file.txt
จากนั้นปิดไฟล์ dummy_file.txt และแสดงสถานะการเปิด/ปิดไฟล์ dummy_text.txt

```
Name of file: dummy_file.txt
Close or not: True
```




````python
#ไม่ใช้คำสั่ง with
dummy = open('dummy_file.txt','r')
print(f'Name of file: {dummy.name}')
dummy.close()
print(f'Close or not: {dummy.closed}')
````

````python
#ใช้คำสั่ง with
with open('dummy_file.txt','r') as dummy:
  print(f'Name of file: {dummy.name}')
  print(f'Close or not: {dummy.closed} (นอก with)')
print(f'Close or not: {dummy.closed} (ใน with)')
````

เขียนโค้ดเพื่อเปิดไฟล์ dummy_file.txt และอ่านเนื้อหาทั้งหมดในไฟล์ เพื่อแสดงผลลัพธ์บนหน้าจอ จากนั้นปิดไฟล์

````python
#ไม่ใช้คำสั่ง with
dummy = open('dummy_file.txt', 'r')
print(dummy.read())
dummy.close()
````

````python
#ใช้คำสั่ง with
with open('dummy_file.txt', 'r') as dummy:
  print(dummy.read())
````
