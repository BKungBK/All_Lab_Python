2. ตัวแปลงอุณหภูมิ
วัตถุประสงค์: ทำงานกับการรับข้อมูลเข้า การแสดงผล และคำสั่งเงื่อนไข

  คำแนะนำ:<br>
  ให้ผู้ใช้ใส่อุณหภูมิในหน่วยเซลเซียส<br>
  แปลงอุณหภูมิเป็นฟาเรนไฮต์<br>
  หากอุณหภูมิต่ำกว่า 0°C ให้พิมพ์ว่า "มันหนาวมากซาร่า!"


```
    กรุณาใส่อุณหภูมิในหน่วยเซลเซียส: -5
    มันหนาวมากซาร่า!
    อุณหภูมิในหน่วยฟาเรนไฮต์: 23.0
```

```
    กรุณาใส่อุณหภูมิในหน่วยเซลเซียส: 35
    อุณหภูมิในหน่วยฟาเรนไฮต์: 95.0
```



````python
C = float(input("กรุณาใส่อุณหภูมิในหน่วยเซลเซียส: "))

F = (C * 1.8) + 32
if C < 0:
  print("มันหนาวมากซาร่า!")
print(f"อุณหภูมิในหน่วยฟาเรนไฮต์: {F}")
````

3. EZ Calculator<br>
วัตถุประสงค์: ใช้ if-elif-else เพื่อจัดการเงื่อนไขหลายๆ อย่าง

  คำแนะนำ:<br>
  1. ให้ผู้ใช้ใส่ตัวเลขสองตัว<br>
  2. ให้ผู้ใช้เลือกการดำเนินการ (+ - * /)<br>
  3. ทำการคำนวณและพิมพ์ผลลัพธ์<br>
  
  Challenge(ทำหรือไม่ทำก็ได้): หากผู้ใช้เลือกเครื่องหมายหาร ให้ตรวจสอบเพิ่มเติม หากตัวหารมีค่า<=0 ให้แสดงข้อความ "Error"


```
    Enter 1st number: 5.6
    Enter 2nd number: 3.2
    Select operator + - * / :+
    Result: 8.8
```


```
    Enter 1st number: 8.5
    Enter 2nd number: 2
    Select operator + - * / :*
    Result: 17.0
```



```
    Enter 1st number: 5
    Enter 2nd number: 2
    Select operator + - * / :a
    Result: wrong operator
```




````python
num_1 = float(input("Enter 1st number: "))
num_2 = float(input("Enter 2nd number: "))
op = input("Select operator + - * / : ")

if op == "+":
  print(f"Result: {num_1 + num_2}")
elif op == "-":
  print(f"Result: {num_1 - num_2}")
elif op == "*":
  print(f"Result: {num_1 * num_2}")
elif op == "/":
  print(f"Result: {num_1 * num_2}")
else:
  print("Result: wrong operator")
````

 4. ตรวจสอบความแข็งแรงของรหัสผ่าน<br>
  วัตถุประสงค์: ใช้เงื่อนไขร่วมกับการทำงานกับสายอักขระ<br>

  คำแนะนำ:<br>
  ให้ผู้ใช้ใส่รหัสผ่าน<br>
  ตรวจสอบความยาวของรหัสผ่าน (ใช้คำสั่ง len())<br>
  พิมพ์ว่ารหัสผ่านนั้น Weak (น้อยกว่า 6 ตัวอักษร), Moderate (6-10 ตัวอักษร), <br>หรือStrong (มากกว่า 10 ตัวอักษร)



```
    Enter password: 123456
    Moderate
```

```
    Enter password: qwertyinwza
    Strong
```


````python
password = len(input("Enter password: "))

if password > 10:
  print("Strong")
elif password >= 6:
  print("Moderate")
else:
  print("Weak")
````

5. หมวดหมู่อายุ<br>
วัตถุประสงค์: ทำงานกับ if-elif-else เพื่อกำหนดหมวดหมู่

  คำแนะนำ:<br>
  ให้ผู้ใช้ใส่อายุของตนเอง<br>
  พิมพ์หมวดหมู่อายุที่สอดคล้อง:<br>
  เด็ก (0-12)<br>
  วัยรุ่น (13-19)<br>
  ผู้ใหญ่ (20-59)<br>
  ผู้สูงอายุ (60+)



```
    กรุณาใส่อายุของคุณ: 35
    ผู้ใหญ่
```


```
    กรุณาใส่อายุของคุณ: -5
    อายุไม่ถูกต้อง
```





````python
old = int(input("กรุณาใส่อายุของคุณ: "))

if old >= 60:
  print("ผู้สูงอายุ")
elif old >= 20:
  print("ผู้ใหญ่")
elif old >= 13:
  print("วัยรุ่น")
elif old >= 0:
  print("เด็ก")
else:
  print("อายุไม่ถูกต้อง")
````

6. ตรวจสอบฤดูกาลตามเดือน
โจทย์: เขียนโปรแกรมที่รับตัวเลขเดือน (1-12) จากผู้ใช้ แล้วแสดงผลฤดูกาลตามเดือนนั้น เช่น

  เดือน 12, 1, 2: Winter<br>
  เดือน 3, 4, 5: Summer<br>
  เดือน 6, 7, 8: Rainy<br>
  เดือน 9, 10, 11: Fall<br>



```
Enter the month number (1-12): 5
The season is Summer
```



```
Enter the month number (1-12): 11
The season is Autumn
```


```
Enter the month number (1-12): 13
Invalid month
```



````python
month = int(input("Enter the month number (1-12): "))

if month in [12, 1, 2]:
  print("Winter")
elif month in [3, 4, 5]:
  print("Summer")
elif month in [6, 7, 8]:
  print("Rainy")
elif month in [9, 10, 11]:
  print("Fall")
else:
  print("Invalid month")
````
