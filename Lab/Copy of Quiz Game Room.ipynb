# Game Room
Game room เป็นระบบสำหรับจัดการผู้เล่นและเกมในห้องเกมออนไลน์ โดยมีสองคลาสคือ Player และ GameRoom


# คลาส Player


เก็บข้อมูล


*   ชื่อผู้เล่น
*   เลเวล(ตัวเลข)
* คะแนนสะสม(ตัวเลข)
* สถานะของผู้เล่น (active, inactive)

มีเมธอดสำหรับ


*  เพิ่มคะแนนสะสมผู้เล่นตามค่าที่ป้อน
*  เพิ่มระดับของผู้เล่นขึ้นหนึ่งขั้น
*  เปลี่ยนสถานะของผู้เล่น
*  แสดงข้อมูลทั้งหมดของผู้เล่น

# คลาส GameRoom

เก็บข้อมูลลิสต์ของผู้เล่นในห้องทั้งหมด
มีเมธอดสำหรับ


*   เพิ่มผู้เล่นใหม่ในห้อง
*   แสดงข้อมูลของผู้เล่นทั้งหมด
*   เพิ่มระดับให้กับผู้เล่นตามชื่อ
*   เปลี่ยนสถานะของผู้เล่นตามชื่อ และสถานะที่กำหนด
*   แจกคะแนนสะสมให้กับผู้เล่นทุกคนในห้องตามจำนวนกำหนด


สร้างคลาส Player และ GameRoom และเขียนโค้ดเพื่อทำการทดสอบดังนี้:


1. เพิ่มผู้เล่น 3 คน
2. เลื่อนระดับผู้เล่น 1 คน
3. เปลี่ยนสถานะผู้เล่น 1 คนเป็น Inactive
4. แจกคะแนนให้ผู้เล่นทั้งหมด
5. แสดงข้อมูลทั้งหมดของผู้เล่น

```python
class Player:
  def __init__(self, name, level, score, status):
    self.player_name = name
    try:
      self.player_level = int(level)
      self.player_score = int(score)
    except TypeError:
      print("|-----------| X กรุณาใส่ตัวเลข X |-----------|")
      return None
    self.player_status = status

  def ScoreUP(self, amount):
    self.player_score += amount
    print(f"|-----------| {self.player_name} คะแนนสะสมเพิ่มขึ้น! {amount} |-----------|")

  def LevelUP(self):
    self.player_level += 1
    print(f"|-----------| Level UP! {self.player_name} |-----------|")

  def ChangeStatus(self, status):
    if status not in ["active", "inactive"]:
      print('|-----------| X กรุณากรอก ["active" or "inactive"] X |-----------|')
      return None
    self.player_status = status
    print(f'|-----------| {self.player_name} มีสถานะเป็น {self.player_status} |-----------|')

  def ShowPlayer(self):
    print(f"|-----------| Status {self.player_name} |-----------|\n Name : {self.player_name}\n Level : {self.player_level}\n Score : {self.player_score}\n Status : {self.player_status}\n|-----------| ------ |-----------|\n")

class GameRoom:
  def __init__(self):
    self.Players = []

  def AddPlayer(self, player):
    self.Players.append(player)
    print(f"|-----------| NewPlayer! |-----------|\n Name : {player.player_name}\n Level : {player.player_level}\n Score : {player.player_score}\n Status : {player.player_status}\n|-----------| ------ |-----------|\n")

  def ShowRoom(self):
    print(f"|----------------------| ALL PLAYER! |----------------------|\n")
    for player in self.Players:
      player.ShowPlayer()
    print(f"|----------------------| ----------- |----------------------|\n")

  def PlyerLevelUP(self, name):
    for player in self.Players:
      if name == player.player_name:
        player.LevelUP()
        return None
    print(f"|-----------| X No Player X |-----------|")
    return None

  def ActiveStatus(self, name, status):
    for player in self.Players:
      if name == player.player_name:
        player.ChangeStatus(status)

  def FreeScore(self, amount):
    for player in self.Players:
      player.ScoreUP(amount)
```

```python
g = GameRoom()

p1 = Player("a", 1, 100, "inactive")
p2 = Player("b", 2, 200, "inactive")
p3 = Player("c", 3, 300, "active")

g.AddPlayer(p1)
g.AddPlayer(p2)
g.AddPlayer(p3)

g.PlyerLevelUP("c")

g.ActiveStatus("a", "inactive")

g.FreeScore(100)

g.ShowRoom()
```
