1.จงเขียนโปรแกรมนับเลขจาก 0 ถึง 5 โดยใช้ for loop และ while loop
```
STARTS
counting 0
counting 1
counting 2
counting 3
counting 4
counting 5
STOP
```



````python
#while loop
num = 0
while num <= 5 :
  if num == 0:
    print("STARTS")
  print(f"counting {num}")
  num += 1
  if num > 5 :
    print("STOP")
````

````python
#for loop
for i in range(6):
  if i == 0:
    print("STARTS")
  print(f"counting {i}")
  if i == 5 :
    print("STOP")
````

4.เขียนโปรแกรมเพื่อรับค่าตัวเลขจากผู้ใช้ จากนั้น คำนวณ summation ตั้งแต่ 1 ถึง ค่าที่ผู้ใช้ป้อน (ใช้ for หรือ while ก็ได้) เช่น ผู้ใช้ป้อนค่า 5 โปรแกรมจะคำนวณผลลัพธ์ของ 1 + 2 + 3 + 4 + 5
```
Enter any number: 5
result : 15
```
```
Enter any number: 3
result : 6
```



````python
sum = 0
r = int(input("Enter any number: "))
for i in range(1,r+1):
  sum += i
print(f"result : {sum}")
````

5.จากโค้ดที่กำหนดให้ จงแก้ไขให้ไม่เกิดลูปอนันต์ (infinite loop)

````python
count = 7
while(count <10):
  if(count < 10):
    print('Count is', count)
````

````python
count = 7
while(count < 10):
  if(count < 10):
    print('Count is', count)
  count += 1
````

6.จงสร้าง List ชื่อ movies เก็บข้อมูลหนังหรือซีรี่ส์ที่นักศึกษาชอบจำนวนห้าเรื่อง จากนั้นใช้ for loop เพื่อแสดงชื่อหนังทั้งห้าเรื่องทางหน้าจอ

````python
l_m = ["Avatar", "Hello World", "JJK", "NOWHERE", "Skibidi"]
for i in l_m:
  print(i)
````

7.จากโปรแกรมก่อนหน้า แก้ไขให้แสดงเฉพาะชื่อของหนังหรือซีรี่ส์ที่มีความยาวมากกว่า 7 ตัวอักษร (ใช้คำสั่ง len()ในการหาจำนวนตัวอักษรของชื่อหนัง)

````python
l_m = ["Avatar", "Hello World", "JJK", "NOWHERE", "Skibidi"]
for i in l_m:
  if len(i) > 7:
    print(i)
````
