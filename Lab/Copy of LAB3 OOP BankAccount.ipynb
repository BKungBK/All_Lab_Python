# Part 2: Designing the BankAccount Class

Task: Refactor and create a `BankAccount` class with the following:


*   Attributes: `account_name, account_balance, account_password`
*   Methods:
    *   `__init__` to initialize account details.
    *   `deposit` to add money.
    *   `withdraw` to remove money if the password matches.
    *   `show_details` to display account information.
    *   add method `receive` to receive money from other accounts without giving password.





```python
class BankAccount:
  def __init__(self, name, balance, password):
    self.ac_name = name
    self.ac_balance = balance
    self.ac_password = password

  def check_password(self, password):
    if password == self.ac_password:
      return True
    else:
      print(" รหัสผ่านของคุณไม่ถูกต้อง ")
      return False

  def deposit(self, count, password):
    if self.check_password(password):
      if count > 0:
        self.ac_balance += count
        print(f"ฝากเงินจำนวน {count} แล้ว\nยอดเงินในบัญชีคงเหลือ : {self.ac_balance}")
      else:
        return print(" ระบุตัวเลขที่ถูกต้อง ")

  def withdraw(self, count, password):
    if self.check_password(password):
      if count > 0 and count < self.ac_balance:
        self.ac_balance -= count
        print(f"ถอนเงินจำนวน {count} แล้ว\nยอดเงินในบัญชีคงเหลือ : {self.ac_balance}")

  def show_details(self):
    print(f"[ Account ]\nname : {self.ac_name}\nbalance : {self.ac_balance}\npassword : {self.ac_password}")

  def show_balance(self):
    print(f"{self.ac_name} มียอดคงเหลือในบัญชี : {self.ac_balance}")

  def receive(self, count):
    print(" รับเงินแล้ว ")
    self.ac_balance += count
```

```python
class Bank:
  def __init__(self):
    self.Account = {}

  def Add_Account(self, account_name, initial_balance, password):
    if account_name in self.Account:
      return print("XXXXXXX มีชื่อบัญชีนี้แล้วกรุณาลองใหม่ในภายหลัง XXXXXXX")
    self.Account.update({account_name :
        BankAccount(account_name, initial_balance, password)
    })
    print(f"[ Create Account! ]\nname : {account_name}\nbalance : {initial_balance}\npassword : {password}\n=================")

  def Check_Password(self, name, password):
    bank_account = self.Account.get(name)
    if name in self.Account:
      if password == bank_account.ac_password:
        return True
      else:
        print("XXXXX รหัสผ่านของคุณไม่ถูกต้อง XXXXX")
        return False
    else:
      print("XXXXX ไม่พบชื่อบัญชีนี้ในระบบ XXXXX")
      return False

  def get_account(self, account_name):
    for i in self.Account:
      if account_name in self.Account:
        bank_account = self.Account.get(account_name)
        print(f"[ Found ! ]")
        print(f"[ Account ]\nname : {account_name}\nbalance : {bank_account.ac_balance}\npassword : {bank_account.ac_password}")
        return None
    print(f" [ NOT Found ! ]\n")

  def deposit(self, name, amount, password):
    bank_account = self.Account.get(name)
    bank_account.deposit(amount, password)

  def withdraw(self, name, amount, password):
     bank_account = self.Account.get(name)
     bank_account.withdraw(amount, password)

  def transfer(self, from_account_name, to_account_name, amount, password):
    bank_account = self.Account.get(from_account_name)
    if self.Check_Password(from_account_name, password):
      if amount > 0 and amount < bank_account.ac_balance:
        bank_account.ac_balance -= amount
        bank_account = self.Account.get(to_account_name)
        bank_account.ac_balance += amount
        print(f">>>>> ได้ทำการโอนเงินจาก {from_account_name} ไปยัง {to_account_name} โอนไป {amount} <<<<<")
      else:
        return print("XXXXX ระบุตัวเลขที่ถูกต้อง XXXXX")

  def bank_summary(self):
    print("[ ALL Account ]\n===============================")
    for i in self.Account:
      bank_account = self.Account.get(i)
      print(f" Account [ {i} ] \nname : {i}\nblance : {bank_account.ac_balance}\npassword : {bank_account.ac_password}\n")
    print("===============================")

b = Bank()
b.bank_summary()
```
