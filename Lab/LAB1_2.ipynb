12. จาก Dictionary ที่กำหนดให้ จงแก้ไขโปรแกรมเพื่อนำสมาชิกในdictionaryมาแสดงผลลัพธ์ดังนี้

Harry Potter is 19 years old student from Gryffindor who masters Defence Against Dark Arts.

```python
#Creating a dictionary
studentInfo = {
    "name" : "Harry Potter",
    "age" : 19,
    "house" : "Gryffindor",
    "master" : "Defence Against Dark Arts"
}

print(f'{studentInfo["name"]} is {studentInfo["age"]} years old student from {studentInfo["house"]} who masters {studentInfo["master"]}')
```

13. จงประกาศ Dictionary ชื่อ address ที่สามารถเก็บข้อมูลบ้านเลขทื่ (houseNo) ถนน (street) เมือง (city) และประเทศ (country) พร้อมทั้งกำหนดค่าเริ่มต้นและแสดงค่าสมาชิกทั้งหมดออกทางหน้าจอ

17. เขียนโปรแกรมเพื่อรับค่ารัศมีวงกลมชนิดทศนิยม จากนั้นคำนวนผลลัพธ์ พื้นที่วงกลม และเส้นรอบวง

ตัวอย่างผลลัพธ์

Enter radius: 2.5

Area = 19.6349375

Circumference = 15.700000000000001



```python
radius = float(input("Enter radius: "))
Area = 3.14 * radius**2
Circumference = 2 * (3.14*radius)

print(f"Area = {Area}\nCircumference = {Circumference}")
```
