```python
class Phone():
  def __init__(self, phone_number):
    self.phone_number = phone_number

  def call_by_number(self, target_number):
    print("Making phone call to ", target_number)

class SmartPhone(Phone):
  def __init__(self, phone_number, email):
    self.email = email
    super().__init__(phone_number)

  def Take_Photo(self):
    print("แชะ!")

  def use_app(self):
    print("app start..")

class CordPhone(Phone):
  def __init__(self, phone_number, color):
    super().__init__(phone_number)
    self.color = color
    self.target_number = ''

  def Diall(self, num):
    self.target_number += num

  def show(self):
    print(self.target_number)

apple = SmartPhone("01245", "bk@gmail.com")
apple.call_by_number("01245")
c = CordPhone("01245", "red")
c.call_by_number("01245")
c.Diall("9")
c.Diall("1")
c.Diall("1")
c.show()
```

```python
class Novice:
  def __init__(self, hp, mp):
    self.hp = hp
    self.mp = mp

  def attack(self):
    print("attack 10 damage")

  def first_aid(self):
    self.mp -= 5
    self.hp += 10
    print(f"Hp : {self.hp}")

class Sword_Man(Novice):
  def __init__(self, hp, mp, STR):
    super().__init__(hp, mp)
    self.STR = STR

  def sword_mastery(self):
    self.mp -= 10
    self.STR *= 2
    print("Use Sword Mastery")

  def bash(self):
    self.mp -= 10
    print(f"Bash {self.STR*10} Damage!")

s = Sword_Man(10, 50, 20)
s.sword_mastery()
s.bash()
```
