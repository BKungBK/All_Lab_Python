```python
weight = float(input("น้ำหนักของคุณ(kg) คือ "))
hight = float(input("ส่วนสูงของคุณ(m) คือ "))

if weight > 300 or weight <= 0:
  print("มั่วละ (น้ำหนักไม่ถูกต้อง)")

if hight > 3 or hight <= 0:
  print("เอาดีๆ (ส่วนสูงไม่ถูกต้อง)")
if hight < 3 and hight > 0 and weight < 300 and weight > 0:
  print(f"bmi ของคุณคือ {weight / hight ** 2}")
```

```python
sc = int(input("Enter Calculus Score :) "))

if sc > 100 or sc < 0:
  print("กรุณากรอกให้ดีด้วยครับ อันนี้มั่วนะ")
elif sc >= 80:
  grade = 'A'
  print(f"Your สุดโหด\nGrade : {grade}")
elif sc >= 70 and sc <= 79:
  grade = 'B'
  print(f"Your ก็ดี\nGrade : {grade}")
elif sc >= 60 and sc <= 69:
  grade = 'C'
  print(f"Your งั้นๆ\nGrade : {grade}")
elif sc >= 50 and sc <= 59:
  grade = 'D'
  print(f"Your ไก่\nGrade : {grade}")
else:
  grade = 'F'
  print(f"ไอกากเอ้ย\nGrade : {grade}")
```
