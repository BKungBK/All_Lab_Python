```python
import time

class LightSwitch():
  def __init__(self, status, color, lumin):
    self.switchIs = status
    self.color = color
    self.lumin = lumin

  def On(self):
    self.switchIs = True

  def Off(self):
    self.switchIs = False

  def changeColor(self, color):
    self.color = color

  def changeLumin(self, amount):
    self.lumin = amount

  def DisplayShow(self):
    print(f"ค่า switchIs : {self.switchIs}\nค่า color : {self.color}\nค่า lumin : {self.lumin} %\n")

Neon = LightSwitch(False, "Red", 100)
Neon2 = LightSwitch(True, "Green", 50)
Neon.DisplayShow()
Neon2.DisplayShow()

class PattaneeBomb():
  def __init__(self, count):
    self.count = count

  def Start(self):
    for i in range(self.count, 0, -1):
      print(i)
      time.sleep(1)

Bomb = PattaneeBomb(10)
Bomb.Start()
print("อ้ากกกกกกกกกกกก !!!!!!!!!!!!!!!!\n อัลเลาะะะะะะะะะ")
```
