```python
class Pokemon:
  def __init__(self, p_name, p_type, p_level, p_health):
    self.__name = p_name
    self.__ptype = p_type
    self.__level = p_level
    self.__health = p_health
    self.__chance = random.randint(40,100)

  def get_name(self):
    return self.__name

  def get_ptye(self):
    return self.__ptype

  def get_level(self):
    return self.__level

  def get_health(self):
    return self.__health

  def get_chance(self):
    return self.__chance

  def set_name(self, new_name):
    self.__name = new_name
    return

  def set_ptye(self, new_ptype):
    self.__ptype = new_ptype
    return

  def set_level(self, new_level):
    self.__level = new_level
    return

  def set_health(self, new_health):
    self.__health = new_health
    return

  def set_chance(self, new_chance):
    self.__chance = new_chance
    return

  def __damage_cal(self):
    __damage = self.__level * 5
    return __damage

  def attack(self, target):
    target.take_damage(self.__damage_cal())
    print(f"{self.get_name()} attacks {target.get_name()}!")

  def take_damage(self, damage):
    self.__health -= damage
    print(f"{self.get_name()} takes {damage} damage! HP left: {self.get_health()}")

  def level_up(self):
    self.__level += 1
    self.__health += 10
    print(f"{self.get_name()} leveled up to {self.get_level()}!")
```

```python
import random

class Pokedex:
  def __init__(self):
    self.__pokemons = []

  def free_Pokemon(self, num):
    if num.lower() in ["1", "pikachu"]:
      self.__pokemons.append(Pokemon('PikachuEx','lightning', 20, 100))
    elif num.lower() in ["2", "mafan"]:
      self.__pokemons.append(Pokemon('Manfan','Dark', 20, 100))

  def get_pokemons(self):
    return self.__pokemons

  def set_pokemons(self, new_pokemons):
    self.__pokemons = new_pokemons

  def show_Pokemons(self):
    print("\n[ You Have ]")
    for pokemon in self.__pokemons:
      print(f"> {pokemon.get_name()}")

  def check_Pokemon(self, p_name):
    for pokemon in self.__pokemons:
      if pokemon.name == p_name:
        return pokemon
      else:
        print("No Pokemon")
        return None

  def catch(self, pokemon, trainer_name):
    if random.choices([True, False],weights = [pokemon.get_chance(), 100 - pokemon.get_chance()],k = 1) == [True]:
      if pokemon not in self.__pokemons:
        self.__pokemons.append(pokemon)
        print(f"{trainer_name} caught {pokemon.get_name()}!")
      else:
        print(f"{trainer_name} have {pokemon.get_name()} in pokeden!")
    else:
      print("Haha loser")
      return

  def release(self, pokemon_name):
    if self.check_Pokemon() != None:
      pokemon = self.check_Pokemon(pokemon_name)
      self.__pokemons.remove(pokemon)
      print(f"{self.name} released {pokemon.get_name()}.")
    else:
      print(f"{pokemon_name} is not in {self.name}'s Pokedex.")
```

```python
class Trainer:
  def __init__(self, T_name):
    self.name = T_name
    self.pokedex = Pokedex()
    self.pokedex.free_Pokemon(input("กรุณาเลือกโปเกม่อนเริ่มต้นของคุณ \n1-Pikachu\n2-Mafan\n>> "))

  def showPokemon(self):
    self.pokedex.show_Pokemons()

  def catch_pokemon(self, pokemon):
    self.pokedex.catch(pokemon, self.name)

  def train_pokemon(self, pokemon_name):
    if self.pokedex.check_Pokemon(pokemon_name) != None:
      pokemon = self.pokedex.check_Pokemon(pokemon_name)
      pokemon.level_up()
      print(f"{self.name} trained {pokemon.get_name()}!")
    else:
      print("No Pokemon")
      return

  def release_pokemon(self, pokemon_name):
    self.pokedex.release(pokemon_name)
```
