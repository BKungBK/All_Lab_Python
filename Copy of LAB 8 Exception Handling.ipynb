จงเขียนโปรแกรมเพื่อรับค่าแรงดันไฟฟ้า V (ทศนิยม) และ ความต้านทานไฟฟ้า R (ทศนิยม) จากนั้นคำนวณกระแสไฟฟ้า I จากสูตร I = V/R โดยใช้คำสั่ง try except เพื่อจัดการ exception ดังต่อไปนี้:

ZeroDivisionError แสดงข้อความ Cannot Devide by Zero (กรณีRเป็นศูนย์)

ValueError แสดงข้อความ Please Enter number(กรณีผู้ใช้ไม่ได้ป้อนค่าตัวเลข)

```python
try:
  V = float(input("Enter V : "))
  R = float(input("Enter R : "))
  print(f"I = {V/R}")
except ZeroDivisionError:
  print("Cannot Devide by Zero")
except ValueError:
  print("Please Enter number")
```

จงใช้คำสั่ง try - except เพื่อจัดการ exception ที่เกิดขึ้น พร้อมแสดงข้อความแจ้งเตือนที่เหมาะสม

```python
try:
  my_list = [1, 2]
  print(f"Third element: {my_list[2]}")
except IndexError:
  print("Error: List index out of range")
```

จงแก้ไขโค้ดต่อไปนี้โดยใช้คำสั่ง raise เพื่อสร้าง exception ชนิด ValueError ที่แสดงข้อความ Invalid Number หากผู้ใช้ป้อนค่าตัวเลขติดลบมากกว่าหรือเท่ากับ 0

```python
num = int(input("Enter negative number: "))

if num < 0:
  raise ValueError("Invalid Number")

print(f"Your number is {num}")
```

เขียนโปรแกรมที่รับความยาวฐานและความสูงของสามเหลี่ยมจากผู้ใช้ แล้วคำนวณหาพื้นที่ของสามเหลี่ยม (สูตร: พื้นที่ = 1/2 × ฐาน × สูง) โดยใช้ try-except เพื่อจัดการข้อผิดพลาดจากการป้อนข้อมูลที่ไม่ใช่ตัวเลข (ValueError) หากคำนวณสำเร็จให้ใช้ else เพื่อแสดงผลลัพธ์ และใช้ finally เพื่อพิมพ์ข้อความว่าการทำงานของโปรแกรมเสร็จสิ้น

ข้อกำหนด:

* รับค่า "ฐาน" และ "ความสูง" ของสามเหลี่ยมจากผู้ใช้
ใช้ try-except
เพื่อป้องกันข้อผิดพลาดจากการป้อนข้อมูลที่ไม่ใช่ตัวเลข (ValueError)

* ใช้ else เพื่อแสดงผลลัพธ์การคำนวณพื้นที่

* ใช้ finally เพื่อแจ้งว่าโปรแกรมทำงานเสร็จสิ้น

```python
try:
  base = float(input("base = "))
  hight = float(input("base = "))
  P = 1/2 * base * hight
except ValueError:
  print("กรุณาใส่ค่าให้ถูกต้อง")
else:
  print(f"Result is {P}")
finally:
  print("End Program")
```
