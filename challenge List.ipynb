```python
def CreateAccount(username, balance, password):
  global Account
  user = getUserAccount(username)
  if user == True:
    print("\nXXXXXX มีชื่อบัญชีนี้ในระบบแล้ว XXXXXX\n")
    return
  Account.append([username, balance, password])
  print(f'\n========== สร้างสำเร็จ ==========\nName : {username}\nBalance {balance}\nPassword : {password}\n=============================\n')

def ShowAccount():
  global Account
  print("\n==================== บัญชีทั้งหมด ====================\n")
  for i in Account:
    print(f'====================\nName : {i[0]}\nBalance {i[1]}\nPassword : {i[2]}')
  print("====================\n\n===================================================\n")

def getUserAccount(username):
  global Username , Account
  for i in Account:
    if i[0] == username:
      Username = list(i)
      return True
  return False

def Login(username, password):
  global _Login
  if getUserAccount(username) == False:
    print("XXXX ไม่มีชื่อนี้ในระบบ XXXX")
    return None
  if Username[2] == password:
    print("\n>>>> Login สำเร็จ! <<<<")
    _Login = True
    print(f"\n>>>> ยินดีต้อนรับคุณ {username} <<<<")
    return None
  else:
    print("\nXXXXXX รหัสผ่านไม่ถูกต้อง XXXXXX\n")
    return None

def getBalance():
  global Username
  return print(f">>>> ยอดเงินคงเหลือ : {Username[1]} บาท. <<<<")

def Deposits(amount):
  if amount <= 0:
    return print("XXXX กรุณาระบุจำนวนเงินที่ถูกต้อง XXXX")
  Username[1] += amount
  return print(f'>>>> ฝากเงินจำนวน : {amount} บาท. <<<<\n>>>> ยอดคงเหลือในบัญชี : {Username[1]} บาท. <<<<')

def Withdraw(amount):
  if amount <= 0:
    return print("XXXX กรุณาระบุจำนวนเงินที่ถูกต้อง XXXX")
  elif amount > Username[1]:
    return print("XXXX กรุณาระบุจำนวนเงินที่ถูกต้อง XXXX")
  Username[1] -= amount
  return print(f'>>>> ถอนเงินจำนวน : {amount} บาท. <<<<\n>>>> ยอดคงเหลือในบัญชี : {Username[1]} บาท. <<<<')
```
