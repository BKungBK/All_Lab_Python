```python
import random
import time

pokemon = {
    "Player": {
        "name" : "",
        "Hp" : 0,
        "Damage": 0,
        "Element": ""
    },
    "Enemy": {
        "name" : "",
        "Hp" : 0,
        "Damage": 0,
        "Element": ""
    }
}

pokemon_element = ["‡∏î‡∏¥‡∏ô","‡∏ô‡πâ‡∏≥","‡∏•‡∏°","‡πÑ‡∏ü","‡∏Å‡∏•‡πâ‡∏≤‡∏°‡∏≠‡∏≤‡∏à‡∏≤‡∏£‡∏¢‡πå‡∏ä‡∏ô"]
enemy_name = ["‡∏≠‡∏≤‡∏à‡∏≤‡∏£‡∏¢‡πå‡∏ä‡∏ô","Skibidi","‡∏°‡∏≤‡πÄ‡∏Ñ‡∏ß‡∏µ‡∏¢","BEE","JUM",]
pokemon_skill = ["‡πÄ‡∏ö‡πà‡∏á‡∏Å‡∏•‡πâ‡∏≤‡∏°","‡∏Å‡∏£‡∏∞‡πÇ‡∏î‡∏î‡∏Å‡∏±‡∏î‡∏Ñ‡∏≠","‡∏ô‡∏¥‡∏û‡∏û‡∏≤‡∏ô","‡∏ï‡∏î","‡∏õ‡∏•‡πà‡∏≠‡∏¢‡∏û‡∏•‡∏±‡∏á‡∏ò‡∏≤‡∏ï‡∏∏"]

print("* „ÄÇ ‚Ä¢ Àö Àö Àõ Àö Àõ ‚Ä¢ „ÄÇ* „ÄÇ¬∞ „ÄÇ* „ÄÇ ‚Ä¢ Àö * „ÄÇ ‚Ä¢ Àö Àö Àõ Àö Àõ \n    ‚ó¢‚ó§‚ó¢‚ó§‚ó¢‚ó§‚ó¢‚ó§‚ó¢‚ó§‚ó¢‚ó§‚ó¢‚ó§‚ó¢‚ó§‚ó¢‚ó§‚ó¢‚ó§‚ó¢‚ó§")
print("          ‡∏¢‡∏¥‡∏ô‡∏î‡∏µ‡∏ï‡πâ‡∏≠‡∏ô‡∏£‡∏±‡∏ö‡πÄ‡∏Ç‡πâ‡∏≤‡∏™‡∏π‡πà‡πÄ‡∏Å‡∏°\n          ‚öîPOKEMON FIGHT!‚öî")
print("    ‚ó¢‚ó§‚ó¢‚ó§‚ó¢‚ó§‚ó¢‚ó§‚ó¢‚ó§‚ó¢‚ó§‚ó¢‚ó§‚ó¢‚ó§‚ó¢‚ó§‚ó¢‚ó§‚ó¢‚ó§\n* „ÄÇ ‚Ä¢ Àö Àö Àõ Àö Àõ ‚Ä¢ „ÄÇ* „ÄÇ¬∞ „ÄÇ* „ÄÇ ‚Ä¢ Àö * „ÄÇ ‚Ä¢ Àö Àö Àõ Àö Àõ ")
print("‚ó¢‚ó§ ‡∏û‡∏¥‡∏°‡∏û‡πå Start ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÄ‡∏£‡∏¥‡πà‡∏°‡πÇ‡∏õ‡∏£‡πÅ‡∏Å‡∏£‡∏° ‡∏´‡∏£‡∏∑‡∏≠ End ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏õ‡∏¥‡∏î‡πÇ‡∏õ‡∏£‡πÅ‡∏Å‡∏£‡∏° ‚ó¢‚ó§")
start = input("‚öî >> ")

while start.lower() != "start":
    if start.lower() == "end":
        exit()
    print("‡∏û‡∏¥‡∏°‡∏û‡πå Start ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÄ‡∏£‡∏¥‡πà‡∏°‡πÇ‡∏õ‡∏£‡πÅ‡∏Å‡∏£‡∏° ‡∏´‡∏£‡∏∑‡∏≠ End ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏õ‡∏¥‡∏î‡πÇ‡∏õ‡∏£‡πÅ‡∏Å‡∏£‡∏°")
    start = input("‚öî >> ")

pokemon["Player"]["name"] = input("‚ó¢‚ó§ ‡∏Å‡∏£‡∏∏‡∏ì‡∏≤‡πÉ‡∏™‡πà‡∏ä‡∏∑‡πà‡∏≠‡πÇ‡∏õ‡πÄ‡∏Å‡∏°‡πà‡∏≠‡∏ô‡∏Ç‡∏≠‡∏á‡∏Ñ‡∏∏‡∏ì ‚öî >> ")
print("\n**************************************************")
print(">>>>>>>>> ‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏™‡∏∏‡πà‡∏° Status <<<<<<<<<")
print("**************************************************")

def random_status(who,r,damage_min,damage_max):
    if who == "Enemy":
        pokemon["Enemy"]["name"] = random.choice(enemy_name)
        pokemon["Enemy"]["Hp"] = random.randint(150,300)
        pokemon["Enemy"]["Damage"] = random.randint(damage_min,damage_max)
        pokemon["Enemy"]["Element"] = random.choice(pokemon_element)
        print("\n----------------------------")
        print(f"‡∏ä‡∏∑‡πà‡∏≠ : {pokemon['Enemy']['name']}\n‡πÄ‡∏•‡∏∑‡∏≠‡∏î : {pokemon['Enemy']['Hp']}\n‡∏û‡∏•‡∏±‡∏á‡πÇ‡∏à‡∏°‡∏ï‡∏µ : {pokemon['Enemy']['Damage']} ‡∏´‡∏ô‡πà‡∏ß‡∏¢\n‡∏ò‡∏≤‡∏ï‡∏∏ : {pokemon['Enemy']['Element']}")
        print("----------------------------")

    delay_time = 0.1
    for i in range(6):
        if r == "hp":
            pokemon["Player"]["Hp"] = random.randint(100,200)
            print(f"Hp ‡∏Ç‡∏≠‡∏á‡∏ô‡πâ‡∏≠‡∏á {pokemon['Player']['name']} ‡∏Ñ‡∏∑‡∏≠ < {pokemon['Player']['Hp']} >")
        elif r == "damage":
            pokemon["Player"]["Damage"] = random.randint(damage_min,damage_max)
            print(f"Damage ‡∏Ç‡∏≠‡∏á‡∏ô‡πâ‡∏≠‡∏á {pokemon['Player']['name']} ‡∏Ñ‡∏∑‡∏≠ < {pokemon['Player']['Damage']} >")
        elif r == "element":
            pokemon["Player"]["Element"] = random.choice(pokemon_element)
            print(f"Element ‡∏Ç‡∏≠‡∏á‡∏ô‡πâ‡∏≠‡∏á {pokemon['Player']['name']} ‡∏Ñ‡∏∑‡∏≠ < {pokemon['Player']['Element']} >")
        time.sleep(delay_time)
        delay_time += 0.2
        if i == 5 and r == "hp":
            print(f"\n‡∏¢‡∏¥‡∏ô‡∏î‡∏µ‡∏î‡πâ‡∏ß‡∏¢‡∏ô‡πâ‡∏≠‡∏á‡πÑ‡∏î‡πâ‡πÄ‡∏•‡∏∑‡∏≠‡∏î {pokemon['Player']['Hp']} ‡∏´‡∏ô‡πà‡∏ß‡∏¢ ‚ô™(¬¥‚ñΩÔΩÄ)\n")
        elif i == 5 and r == "damage":
            print(f"\n‡∏¢‡∏¥‡∏ô‡∏î‡∏µ‡∏î‡πâ‡∏ß‡∏¢‡∏ô‡πâ‡∏≠‡∏á‡πÑ‡∏î‡πâ‡∏û‡∏•‡∏±‡∏á‡πÇ‡∏à‡∏°‡∏ï‡∏µ {pokemon['Player']['Damage']} ‡∏´‡∏ô‡πà‡∏ß‡∏¢ ‚ô™(¬¥‚ñΩÔΩÄ)\n")
        elif i == 5 and r == "element":
            print(f"\n‡∏¢‡∏¥‡∏ô‡∏î‡∏µ‡∏î‡πâ‡∏ß‡∏¢‡∏ô‡πâ‡∏≠‡∏á‡πÑ‡∏î‡πâ‡∏ò‡∏≤‡∏ï‡∏∏ {pokemon['Player']['Element']} (‚ïØ‚ñΩ‚ï∞ )\n")

def check_element(player_element, opponent_element):
    if player_element == "‡∏Å‡∏•‡πâ‡∏≤‡∏°‡∏≠‡∏≤‡∏à‡∏≤‡∏£‡∏¢‡πå‡∏ä‡∏ô":
        print(f"‡∏ò‡∏≤‡∏ï‡∏∏ {player_element} ‡∏ä‡∏ô‡∏∞‡∏ó‡∏∏‡∏Å‡∏ò‡∏≤‡∏ï‡∏∏!")
        pokemon["Enemy"]["Damage"] *= 0.90
    elif player_element == opponent_element:
        print(f"‡∏ò‡∏≤‡∏ï‡∏∏ ‡πÄ‡∏´‡∏°‡∏∑‡∏≠‡∏ô‡∏Å‡∏±‡∏ô!")
    elif player_element == "‡∏ô‡πâ‡∏≥" and opponent_element == "‡πÑ‡∏ü":
        print(f"‡∏ò‡∏≤‡∏ï‡∏∏ {player_element} ‡∏ä‡∏ô‡∏∞‡∏ò‡∏≤‡∏ï‡∏∏ {opponent_element}!")
        pokemon["Enemy"]["Damage"] *= 0.90
    elif player_element == "‡πÑ‡∏ü" and opponent_element == "‡∏•‡∏°":
        print(f"‡∏ò‡∏≤‡∏ï‡∏∏ {player_element} ‡∏ä‡∏ô‡∏∞‡∏ò‡∏≤‡∏ï‡∏∏ {opponent_element}!")
        pokemon["Enemy"]["Damage"] *= 0.90
    elif player_element == "‡∏•‡∏°" and opponent_element == "‡∏î‡∏¥‡∏ô":
        print(f"‡∏ò‡∏≤‡∏ï‡∏∏ {player_element} ‡∏ä‡∏ô‡∏∞‡∏ò‡∏≤‡∏ï‡∏∏ {opponent_element}!")
        pokemon["Enemy"]["Damage"] *= 0.90
    elif player_element == "‡∏î‡∏¥‡∏ô" and opponent_element == "‡∏ô‡πâ‡∏≥":
        print(f"‡∏ò‡∏≤‡∏ï‡∏∏ {player_element} ‡∏ä‡∏ô‡∏∞‡∏ò‡∏≤‡∏ï‡∏∏ {opponent_element}!")
        pokemon["Enemy"]["Damage"] *= 0.90
    else:
        print(f"‡∏ò‡∏≤‡∏ï‡∏∏ {player_element} ‡πÅ‡∏û‡πâ‡∏ò‡∏≤‡∏ï‡∏∏ {opponent_element}. ‡∏•‡∏î‡∏î‡∏≤‡πÄ‡∏°‡∏à‡∏Ç‡∏≠‡∏á‡∏ú‡∏π‡πâ‡πÄ‡∏•‡πà‡∏ô‡∏•‡∏á 10%!")
        pokemon["Player"]["Damage"] *= 0.90
    pokemon["Player"]["Damage"] = int(pokemon["Player"]["Damage"])
    pokemon["Enemy"]["Damage"] = int(pokemon["Enemy"]["Damage"])

def math_coke():
    num1 = random.randint(1, 100)
    num2 = random.randint(1, 100)

    operators = ['+', '-']
    operator = random.choice(operators)

    problem = f"{num1} {operator} {num2}"
    result_math = eval(problem)

    return problem,result_math

# ‡∏™‡∏∏‡πà‡∏° Status
print("**************************************************")
print(">>>>>>>>> ‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏™‡∏∏‡πà‡∏° HP <<<<<<<<<")
print("**************************************************")
random_status("","hp",0,0)
time.sleep(1)
print("**************************************************")
print(">>>>>>>>> ‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏™‡∏∏‡πà‡∏° Damage <<<<<<<<<")
print("**************************************************")
random_status("","damage",10,40)
time.sleep(1)
print("**************************************************")
print(">>>>>>>>> ‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏™‡∏∏‡πà‡∏° Element <<<<<<<<<")
print("**************************************************")
random_status("","element",0,0)
time.sleep(1)
print(f"\n ö Ô∏µ‚ÄøÔ∏µ‚ÄøÔ∏µÍí∞ All Status Íí±Ô∏µ‚ÄøÔ∏µ‚ÄøÔ∏µ …û\n‚îä__________________________________‚îä\nÍí∞ üí≠ Íí± Name : {pokemon['Player']['name']}\nÍí∞ üí≠ Íí± Hp : {pokemon['Player']['Hp']} ‡∏´‡∏ô‡πà‡∏ß‡∏¢\nÍí∞ üí≠ Íí± Damage : {pokemon['Player']['Damage']} ‡∏´‡∏ô‡πà‡∏ß‡∏¢\nÍí∞ üí≠ Íí± Element : {pokemon['Player']['Element']}\n__________________________________\nÔ∏∂Ô∏∂Ô∏∂Ô∏∂Ô∏∂Ô∏∂Ô∏∂Ô∏∂Ô∏∂Ô∏∂Ô∏∂Ô∏∂")

print("\n**************************************************")
print("‚ó¢‚ó§ ‡∏´‡∏≤‡∏Å‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡∏î‡∏≥‡πÄ‡∏ô‡∏¥‡∏ô‡∏Å‡∏≤‡∏£‡∏ï‡πà‡∏≠‡∏û‡∏¥‡∏°‡∏û‡πå Go ‡∏´‡∏£‡∏∑‡∏≠ End ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏õ‡∏¥‡∏î‡πÇ‡∏õ‡∏£‡πÅ‡∏Å‡∏£‡∏°")
start_Fight = input("‚öî >> ")
while start_Fight.lower() != "go":
    if start_Fight.lower() == "end":
        exit()
    print("‡∏û‡∏¥‡∏°‡∏û‡πå Go ‡∏´‡∏£‡∏∑‡∏≠ End ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏õ‡∏¥‡∏î‡πÇ‡∏õ‡∏£‡πÅ‡∏Å‡∏£‡∏°")
    start_Fight = input("‚öî >> ")

print("\n##################")
print("‡∏Ñ‡∏∏‡∏ì‡∏à‡∏∞‡∏ï‡πâ‡∏≠‡∏á‡∏™‡∏π‡πâ‡∏Å‡∏±‡∏ö!!!!!")
print("##################")
time.sleep(2)
random_status("Enemy","",10,50)

# ‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏Å‡∏≤‡∏£‡∏ï‡πà‡∏≠‡∏™‡∏π‡πâ
print("\n**************************************************")
print(">>>>> ‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏Å‡∏≤‡∏£‡∏ï‡πà‡∏≠‡∏™‡∏π‡πâ <<<<<<")
print("**************************************************\n")
Drink = input("‡∏´‡∏≤‡∏Å‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡πÉ‡∏´‡πâ‡∏ô‡πâ‡∏≠‡∏á‡∏î‡∏∑‡πà‡∏° Coke ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏û‡∏•‡∏±‡∏á‡πÇ‡∏à‡∏°‡∏ï‡∏µ x3 ‡∏ï‡πâ‡∏≠‡∏á‡∏ï‡∏≠‡∏ö‡πÇ‡∏à‡∏ó‡∏¢‡πå‡∏Ñ‡∏ì‡∏¥‡∏ï‡πÉ‡∏´‡πâ‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á!\n‡∏Å‡∏£‡∏∏‡∏ì‡∏≤‡∏û‡∏¥‡∏°‡∏û‡πå yes ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏î‡∏∑‡πà‡∏°‡∏´‡∏£‡∏∑‡∏≠‡∏û‡∏¥‡∏°‡∏û‡πå no ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÑ‡∏°‡πà‡∏î‡∏∑‡πà‡∏°\n‚öî >> ")
if Drink.lower() == "yes":
    problem,result_math = math_coke()
    print(f"‡πÇ‡∏à‡∏ó‡∏¢‡πå ‡∏Ñ‡∏∑‡∏≠ {problem} = ?")
    result_player = int(input("‡∏Ñ‡∏≥‡∏ï‡∏≠‡∏ö ‡∏Ñ‡∏∑‡∏≠ "))
    if result_player == result_math:
        pokemon["Player"]["Damage"] *= 3
        print("\n**************************************************")
        print(">>>>> ‡∏Ñ‡∏≥‡∏ï‡∏≠‡∏ö‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á! <<<<<<")
        print(f"‡∏ô‡πâ‡∏≠‡∏á {pokemon['Player']['name']} ‡πÑ‡∏î‡πâ‡∏û‡∏•‡∏±‡∏á‡πÇ‡∏à‡∏°‡∏ï‡∏µ‡∏Ç‡∏∂‡πâ‡∏ô‡πÄ‡∏õ‡πá‡∏ô {pokemon['Player']['Damage']}\n")
        print("**************************************************\n")
    else:
        print("‡∏Ñ‡∏≥‡∏ï‡∏≠‡∏ö‡∏ú‡∏¥‡∏î‡∏à‡πâ‡∏≤ ‡πÄ‡∏™‡∏µ‡∏¢‡πÉ‡∏à‡∏î‡πâ‡∏ß‡∏¢\n")
time.sleep(1)
check_element(pokemon["Player"]["Element"],pokemon["Enemy"]["Element"])
time.sleep(2)

# ‡∏ú‡∏•‡∏Å‡∏≤‡∏£‡∏ï‡πà‡∏≠‡∏™‡∏π‡πâ
winner = ""
while pokemon["Player"]["Hp"] >= 0 or pokemon["Enemy"]["Hp"] >= 0:
    pokemon["Enemy"]["Hp"] -= pokemon["Player"]["Damage"]
    if pokemon["Enemy"]["Hp"] <= 0:
        print(f'\n========= ‡∏ú‡∏•‡∏Å‡∏≤‡∏£‡∏ï‡πà‡∏≠‡∏™‡∏π‡πâ =========\n{pokemon["Player"]["name"]} {random.choice(pokemon_skill)}‡πÉ‡∏™‡πà {pokemon["Enemy"]["name"]}\n{pokemon["Enemy"]["name"]} ‡πÄ‡∏•‡∏∑‡∏≠‡∏î‡∏´‡∏°‡∏î‡πÅ‡∏•‡πâ‡∏ß {pokemon["Player"]["name"]} ‡∏ä‡∏ô‡∏∞‡πÑ‡∏õ‡πÅ‡∏ö‡∏ö‡πÉ‡∏™‡πÜ\n============================')
        winner = pokemon["Player"]["name"]
        break
    print(f'‡∏ô‡πâ‡∏≠‡∏á {pokemon["Player"]["name"]} {random.choice(pokemon_skill)}‡πÉ‡∏™‡πà {pokemon["Enemy"]["name"]}\n{pokemon["Enemy"]["name"]} ‡πÄ‡∏•‡∏∑‡∏≠‡∏î‡πÄ‡∏´‡∏•‡∏∑‡∏≠ {pokemon["Enemy"]["Hp"]}\n')
    time.sleep(1)
    pokemon["Player"]["Hp"] -= pokemon["Enemy"]["Damage"]
    if pokemon["Player"]["Hp"] <= 0:
        print(f'\n========= ‡∏ú‡∏•‡∏Å‡∏≤‡∏£‡∏ï‡πà‡∏≠‡∏™‡∏π‡πâ =========\n{pokemon["Enemy"]["name"]} {random.choice(pokemon_skill)}‡πÉ‡∏™‡πà {pokemon["Player"]["name"]}\n{pokemon["Player"]["name"]} ‡πÄ‡∏•‡∏∑‡∏≠‡∏î‡∏´‡∏°‡∏î‡πÅ‡∏•‡πâ‡∏ß {pokemon["Enemy"]["name"]} ‡∏ä‡∏ô‡∏∞‡πÑ‡∏õ‡πÅ‡∏ö‡∏ö‡πÉ‡∏™‡πÜ\n============================')
        winner = pokemon["Enemy"]["name"]
        break
    print(f'‡∏ô‡πâ‡∏≠‡∏á {pokemon["Enemy"]["name"]} ‡πÑ‡∏°‡πà‡∏¢‡∏≠‡∏° {random.choice(pokemon_skill)}‡πÉ‡∏™‡πà {pokemon["Player"]["name"]}\n{pokemon["Player"]["name"]} ‡πÄ‡∏•‡∏∑‡∏≠‡∏î‡πÄ‡∏´‡∏•‡∏∑‡∏≠ {pokemon["Player"]["Hp"]}\n')
if winner == pokemon["Player"]["name"]:
    print("  ‚àß,,,‚àß")
    print("(  Ã≥‚Ä¢ ¬∑ ‚Ä¢ Ã≥)")
    print("/    „Å•(‡∏¢‡∏¥‡∏ô‡∏î‡∏µ‡∏î‡πâ‡∏ß‡∏¢‡∏Ñ‡∏∏‡∏ì‡∏ä‡∏ô‡∏∞‡πÅ‡∏•‡πâ‡∏ß!!)")
    print("‡∏ï‡∏≠‡∏ô‡∏ô‡∏µ‡πâ‡∏Ñ‡∏∏‡∏ì‡πÄ‡∏õ‡πá‡∏ô‡πÄ‡∏à‡πâ‡∏≤‡∏Ç‡∏≠‡∏á‡πÇ‡∏•‡∏Å‡πÅ‡∏•‡πâ‡∏ß‡∏Ç‡∏≠‡πÉ‡∏´‡πâ‡∏°‡∏µ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏∏‡∏Ç‡∏ô‡∏∞")
else:
    print("‡∏Ñ‡∏∏‡∏ì‡∏ô‡∏µ‡πâ‡∏°‡∏±‡∏ô loser ‡∏ï‡∏±‡∏ß‡∏û‡πà‡∏≠‡∏à‡∏£‡∏¥‡∏á‡πÜ")
print("\n‚âΩ ^ ‚Ä¢ ‚©ä ‚Ä¢ ^ ‚âº\n‚û§ ‡∏ó‡∏≥‡∏Å‡∏≤‡∏£‡∏õ‡∏¥‡∏î‡πÇ‡∏õ‡∏£‡πÅ‡∏Å‡∏£‡∏°‡∏≠‡∏±‡∏ï‡∏¥‡πÇ‡∏ô‡∏°‡∏±‡∏ï‡∏¥")
exit()
```
