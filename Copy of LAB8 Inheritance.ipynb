# 1: Single Inheritance
ทดสอบความเข้าเรื่อง superclass และ subclass โดยให้นักศึกษาสร้างคลาส Vehicle โดยมีคุณสมบัติดังนี้

Attribute:
*   brand สำหรับเก็บยี่ห้อ
*   speed สำหรับเก็บความเร็ว

Method:
*   display_info() แสดงยี่ห้อรถ และความเร็วในหน่วย km/h
*   accelerate() เพิ่มความเร็วของรถตามจำนวนที่ระบุ และแสดงความเร็วใหม่ทางหน้าจอ

จากนั้นสร้างซับคลาส Car โดยสืบทอดจากคลาส Vehicle โดยมี Attribute เพิ่มคือ

*   fuel_type เพื่อเก็บชนิดของเชื้อเพลิง

และมีเมธอดเพิ่มคือ

*   apply_brake() ซึ่งลดความเร็วของรถตามจำนวนที่ระบุ โดยความเร็วของรถห้ามเป็นค่าติดลบ

จากนั้นแก้ไข(overide)การแสดงผลของเมธอด display_info()ของ class Vehicle เพื่อเพิ่มส่วนของการแสดงผล fuel_type


```python
class Vehicle:
  def __init__(self, brand, speed):
    try:
      self.__brand = str(brand)
      self.__speed = float(speed)
    except ValueError:
      #เรียกใช้งานได้แต่ก็ข้อมูลเป็นค่าเริ่มต้น
      self.__brand = 'Unknow'
      self.__speed = 0
      print("กรุณากรอกข้อมูลให้ถูกต้อง!")
      return

  def get_brand(self):
    return self.__brand

  def set_brand(self, new):
    self.__brand = str(new)

  def get_speed(self):
    return self.__speed

  def set_speed(self, new):
    self.__speed = max(0, new)

  def display_info(self):
    print(f"========== {self.__brand} ==========")
    #ผมไม่รู้ว่าเขาเอาทศนิยมกันไหมแต่ไม่ชอบที่มี . 5555
    print(f" Speed : {int(self.__speed)} km/h")
    print(f"========== {'=' * len(self.__brand)} ==========")

  def accelerate(self, amount):
    try:
      self.__speed += float(amount)
      print(f"========== New Speed! ==========")
      print(f" Speed : {int(self.__speed)} km/h")
      print(f"========== ========== ==========")
    except ValueError:
      print("กรุณากรอกข้อมูลให้ถูกต้อง!")
      return

class Car(Vehicle):
  def __init__(self, brand, speed, fuel_type):
    super().__init__(brand, speed)
    self.__fuel_type = str(fuel_type)

  def get_fuel_type(self):
    return self.__fuel_type

  def set_fuel_type(self, new):
    self.__fuel_type = str(new)

  def apply_brake(self, amount):
    try:
      self.set_speed(max(0 ,(self.get_speed() - float(amount))))
      print(f"========== Speed ==========")
      print(f" Speed : {int(self.get_speed())} km/h")
      print(f"========== ========== ==========")
    except ValueError:
      print("กรุณากรอกข้อมูลให้ถูกต้อง!")
      return

  def display_info(self):
    print(f"========== {self.get_brand()} ==========")
    print(f" Speed : {int(self.get_speed())} km/h")
    print(f" Fuel Type : {self.__fuel_type}")
    print(f"========== {'=' * len(self.get_brand())} ==========")
```

สร้างออปเจค Car และทดสอบการเรียกใช้เมธอด

```python
toyota = Car("toyota", 90, "bensin")
toyota.display_info()
toyota.accelerate(20)
toyota.apply_brake(300)
toyota.display_info()
```

# 2: Multilevel Inheritance
สร้างsubclass ElectricCar สืบทอดจากคลาส Car โดยเพิ่มแอตทริบิวต์
* battery_capacity สำหรับเก็บความจุแบตเตอรี่

และเมธอด
* battery_status() เพื่อเช็คสถานะของแบตเตอรี่ หากมากกว่า 90 ให้แสดงสถานะ Full Battery หากอยู่ในช่วง 40-89 แสดงสถานะ Normal battery และหากต่ำกว่า 40 แสดงสถานะ Low battery

จากนั้น overide เมธอด display_info() จากคลาส Car โดยเพิ่มส่วนของการแสดงผล battery_capacity ในหน่วย kWh

```python
class ElectricCar(Car):
  def __init__(self, brand, speed, fuel_type, battery_capacity):
    super().__init__(brand, speed, fuel_type)
    try:
      self.__battery_capacity = float(battery_capacity)
    except ValueError:
      self.__battery_capacity = 0.0
      print("กรุณากรอกข้อมูลให้ถูกต้อง!")
      return

  def get___battery_capacity(self):
    return self.__battery_capacity

  def set_battery_capacity(self, new):
    try:
      self.__battery_capacity = float(new)
    except ValueError:
      print("กรุณากรอกข้อมูลให้ถูกต้อง!")
      return

  def display_battery(self):
    #▰ ▱
    if self.__battery_capacity >= 90:
      status = "Full Battery"
    elif self.__battery_capacity >= 40:
      status = "Normal battery"
    else:
      status = "Low battery"
    battery_percent = "▰" * int(self.__battery_capacity / 10)
    battery_npercent = "▱" * abs(int(self.__battery_capacity / 10) - 10)
    print(f" Battery : {status}")
    print(f"{battery_percent + battery_npercent} {int(self.__battery_capacity)}%".center(20))

  def battery_status(self):
    print(f"========== [Battery] ==========")
    self.display_battery()
    print(f"========== {'=' * len(self.get_brand())} ==========")

  def display_info(self):
    print(f"========== {self.get_brand()} ==========")
    print(f" Speed : {int(self.get_speed())} km/h")
    print(f" Fuel Type : {self.get_fuel_type()}")
    self.display_battery()
    print(f"========== {'=' * len(self.get_brand())} ==========")

e = ElectricCar("toyota", 90, "battery", 30)
e.display_info()
```

#3: Multiple Inheritance

สร้างคลาส GPS เพื่อเก็บพิกัด latitude และ longtitude และมีเมธอด
* display_location() เพื่อแสดงพิกัด
* update_location() เพื่ออัปเดตตำแหน่ง

จากนั้นสร้างคลาส SmartElectricCar สำหรับรถไฟฟ้าอัจริยะโดยคลาส SmartElectricCar สืบทอดจากทั้ง ElectricCar และ GPS
โดยมีเมธอดเพิ่มเติมคือ
* self_driving_mode() เพื่อเปิดโหมดการขับขี่แบบ autopilot โดยแสดงผลข้อความ Autopilot activated หากความเร็วของรถต่ำกว่า 100 มิฉะนั้นจะแสดงข้อความ Manual driving required


จากนั้นoverideเมธอด display_info()โดยเพิ่มการแสดงผลพิกัด GPS ของรถ



```python
class GPS:
  def __init__(self, latitude, longtitude):
    try:
      self.__latitude = float(latitude)
      self.__longtitude = float(longtitude)
    except ValueError:
      self.__latitude = 0
      self.__longtitude = 0
      print("กรุณากรอกข้อมูลให้ถูกต้อง!")
      return

  def get_latitude(self):
    return self.__latitude

  def get_longtitude(self):
    return self.__longtitude

  def set_latitude(self, new):
    try:
      self.__latitude = float(new)
    except ValueError:
      print("กรุณากรอกข้อมูลให้ถูกต้อง!")
      return

  def set_longtitude(self, new):
    try:
      self.__longtitude = float(new)
    except ValueError:
      print("กรุณากรอกข้อมูลให้ถูกต้อง!")
      return

  def display_info(self):
    banner = "•──────────────────•°•°•──────────────────•"
    print(banner)
    print(f"latitude({self.__latitude}) longtitude({self.__longtitude})".center(len(banner)))
    print(banner)

  def update_location(self, new_latitude, new_longtitude):
    try:
      self.__latitude = float(new_latitude)
      self.__longtitude = float(new_longtitude)
      self.display_info()
    except ValueError:
      print("กรุณากรอกข้อมูลให้ถูกต้อง!")
      return

class SmartElectricCar(ElectricCar, GPS):
  def __init__(self, brand, speed, fuel_type, battery_capacity, latitude, longtitude):
    super().__init__(brand, speed, fuel_type, battery_capacity)
    GPS.__init__(self, latitude, longtitude)

  def self_driving_mode(self):
    if int(self.get_speed()) <= 100:
      print("Autopilot activated")
    else:
      print("Manual driving required")

  def display_info(self):
    print(f"========== {self.get_brand()} ==========")
    print(f" Speed : {int(self.get_speed())} km/h")
    print(f" Fuel Type : {self.get_fuel_type()}")
    self.display_battery()
    print(f"GPS : latitude({self.get_latitude()}) longtitude({self.get_longtitude()})")
    print(f"========== {'=' * len(self.get_brand())} ==========")
```

สร้างออปเจคคลาส SmartElectricCar จากนั้นทอสอบการใช้งานเมธอด

```python
s = SmartElectricCar("toyota", 90, "battery", 30, 98, 54)
s.display_info()
```
